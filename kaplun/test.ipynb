```python
class Reader:
    def __call__(self, data_):
        self.data = pd.DataFrame()


class XMLReader(Reader):
    def __init__(self):
        self.data = None
        self.data_bin: bytes = b''
    def __call__(self, data_):
        self.data = pd.read_xml(data_)
        self.data_bin: bytes = data_


# class CSVReader(Reader):
#     def __init__(self):
#         self.data = None
#         self.data_bin = None
#     def __call__(self, data_):
#         self.data = pd.read_csv(data_)
#         self.data_bin: bytes = data_
# r = CSVReader()
```
